```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing data frames
US500=pd.read_csv('SPY Historical Data.csv')
NOBL=pd.read_csv('NOBL Historical Data.csv')
O=pd.read_csv('O Historical Data.csv')
PBA=pd.read_csv('PBA Historical Data.csv')
SCHD=pd.read_csv('SCHD Historical Data.csv')
SDIV=pd.read_csv('SDIV Historical Data.csv')
```

```python
#rename and drop unnecesary columns colums
#US500
US500.rename(columns={'Price':'Price US500'}, inplace=True)
US500.rename(columns={'Change %':'Change % US500'}, inplace=True)
US500.drop(columns=['Open','High','Low','Vol.'],inplace=True)
#NOBL
NOBL.rename(columns={'Price':'Price NOBL'}, inplace=True)
NOBL.rename(columns={'Change %':'Change % NOBL'}, inplace=True)
NOBL.drop(columns=['Open','High','Low','Vol.'],inplace=True)
#O
O.rename(columns={'Price':'Price O'}, inplace=True)
O.rename(columns={'Change %':'Change % O'}, inplace=True)
O.drop(columns=['Open','High','Low','Vol.'],inplace=True)
#PBA
PBA.rename(columns={'Price':'Price PBA'}, inplace=True)
PBA.rename(columns={'Change %':'Change % PBA'}, inplace=True)
PBA.drop(columns=['Open','High','Low','Vol.'],inplace=True)
#SCHD
SCHD.rename(columns={'Price':'Price SCHD'}, inplace=True)
SCHD.rename(columns={'Change %':'Change % SCHD'}, inplace=True)
SCHD.drop(columns=['Open','High','Low','Vol.'],inplace=True)
#SDIV
SDIV.rename(columns={'Price':'Price SDIV'}, inplace=True)
SDIV.rename(columns={'Change %':'Change % SDIV'}, inplace=True)
SDIV.drop(columns=['Open','High','Low','Vol.'],inplace=True)
```

```python
#convert Change % to number
US500['Change % US500']=US500['Change % US500'].astype('string')
US500['Change % US500']=US500['Change % US500'].str.replace("%","").astype(float)
NOBL['Change % NOBL']=NOBL['Change % NOBL'].astype('string')
NOBL['Change % NOBL']=NOBL['Change % NOBL'].str.replace("%","").astype(float)
O['Change % O']=O['Change % O'].astype(pd.StringDtype())
O['Change % O']=O['Change % O'].str.replace("%","").astype(float)
PBA['Change % PBA']=PBA['Change % PBA'].astype('string')
PBA['Change % PBA']=PBA['Change % PBA'].str.replace("%","").astype(float)
SCHD['Change % SCHD']=SCHD['Change % SCHD'].astype('string')
SCHD['Change % SCHD']=SCHD['Change % SCHD'].str.replace("%","").astype(float)
SDIV['Change % SDIV']=SDIV['Change % SDIV'].astype('string')
SDIV['Change % SDIV']=SDIV['Change % SDIV'].str.replace("%","").astype(float)
US500['Change % US500'].dtypes
```

```python
#Create lines,to see the last 10 month of data
y1=US500["Change % US500"].tail(10)
y2=NOBL["Change % NOBL"].tail(10)
y3=O["Change % O"].tail(10)
y4=PBA["Change % PBA"].tail(10)
y5=SCHD["Change % SCHD"].tail(10)
y6=SDIV["Change % SDIV"].tail(10)
plt.figure(figsize=(20,20))
plt.plot(y1, label = "US500",color='blue')
plt.plot(y2, label = "NOBL",color='red')
plt.plot(y3, label = "O",color='yellow')
plt.plot(y4, label = "PBA",color='brown')
plt.plot(y5, label = "SCHD",color='green')
plt.plot(y6, label = "SDIV",color='purple')
plt.legend()
plt.show()
```
